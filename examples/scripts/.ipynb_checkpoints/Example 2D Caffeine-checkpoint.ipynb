```python
orientations = np.load('../simulation/orientations_10000.npy')

convertion_factor = np.array([101*101,101,1],dtype = int)

pixel_num = det.pix_num_total

# Process 100 images a time.
weight_holder=np.zeros((101*101*101))
value_holder=np.zeros((101*101*101))

for l in range(10):
    
    # Get the pixel position
    pixel_position = det.pixel_position_reciprocal.copy()
    
    # Load pattern
    data = np.load('../simulation/raw/rdv_100_random_slices_batch_%d.npy'%l)
    
    tic = time.time()
    
    for m in range(100):
        
        # Rotate the pixel position
        rotmat = ps.geometry.quaternion2rot3D(orientations[l*100+m,:])

        # rotate the pixels in the reciprocal space. Notice that at this time, the pixel position is in 3D
        pixel_position = ps.geometry.rotate_pixels_in_reciprocal_space(rotmat, pixel_position)
        
        # get index and weight
        index ,weight = ps.geometry.get_weight_in_reciprocal_space(pixel_position, voxel_length, 101)

        index_2d = np.reshape(index, [pixel_num, 8, 3])
        index_2d = np.matmul(index_2d, convertion_factor)
        
        weight_2d = np.reshape(weight,[pixel_num, 8])
        
        assign_values(index_2d, weight_2d, np.reshape(data[m,:],pixel_num),
                      pixel_num ,value_holder, weight_holder)
    
    toc = time.time()
    print("It takes %f seconds to merge 100 images"%(toc-tic))
```

```python
@jit
def assign_values(ind_, wei_, val_, num_ ,val_hol_, wei_hol_):
    for l in xrange(num_):
        for m in range(8):
            val_hol_[ind_[l,m]] += val_[l]*wei_[l,m]
            wei_hol_[ind_[l,m]] += wei_[l,m]
```
