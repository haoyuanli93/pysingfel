```python
import numpy as np
import h5py as h5
import matplotlib.pyplot as plt
import pysingfel as ps
import time
from numba import jit
import psana
```

```python
@jit
def Bin4by4_rescale(pattern, shape,scale):
    
    result = np.zeros(shape)
    for l in range(shape[0]):
        for m in range(shape[1]):
            result[l,m] = np.sum(pattern[l*4:l*4+4, m*4:m*4+4])/scale
            
    return result
```

```python
# Load amo86615 related informations to use the detector class
ds = psana.DataSource('exp=amo86615:run=182:idx')
run = ds.runs().next()
times = run.times()
env = ds.env()
evt = run.event(times[0])
exp_det = psana.Detector('pnccdBack', env)

# Load beam
beam = ps.Beam('../input/s2e_readPDB.beam') 

# Load and initialize the detector
det = ps.Detector(mode = "LCLS Detectors")
det.initialize_as_LCLS_detector('../../lcls_detectors/pnccdFront/group/source/geometry')
det.initialize_pixels_with_beam_information(polar=[0,1,0], wavevector = beam.get_wavevector())
background = np.sqrt(3*det.pixel_rms)

# Create variable to holds the downsampled photons
ds_photons = np.zeros((10000,260,256))

# load data
with h5.File('../simulation/photons/photons.h5','r+') as f:
    for l in range(100):
        data = f[str(l)][:]
        for m in range(100):
            adu = data[m,:,:]
            # convert to 2d and subtract the static noise
            adu = det.convert_image_stack_to_2d_image(adu-background)
            adu[adu<=0] = 0
            # Downsampling. Notice that this is a simulation and the shape is good.
            ds_photons[l*100+m,:,:] = Bin4by4_rescale(adu,(260,256), 130.)
        print('Finishes batch %d'%l)    
        
    ds_photons = ds_photons.astype(np.int32)
    f.create_dataset('ds photons',data = ds_photons)
```

```python
with h5.File('../photons/all_adus.h5','r+') as f:
    ds_photons = ds_photons.astype(np.int32)
    f.create_dataset('ds photons',data = ds_photons)
```
