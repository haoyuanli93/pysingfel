# Start the test

```python
import numpy as np
import matplotlib.pyplot as plt
import time

from particle import *
from beam import *
from detector import *
from toolbox import *
from radiationDamage import rotateParticle
```

# Calculate Differential Scattering Section

```python
name = '2'

# Load particle
particle = Particle()
particle.readPDB('../input/DNA.pdb', ff='WK')

quaternion = getRandomRotation('x')
rotateParticle(quaternion, particle)

print "atom number is:", particle.SplitIdx[-1]

# Load beam
beamFile = '../input/s2e_readPDB.beam'
beam = Beam(beamFile) 

# Load and initialize the detector
det = Detector()
det.initialize_as_LCLS_detector('../../lcls_detectors/pnccdFront/group/source/geometry/')
Wavevector = np.array([0,0,beam.get_wavenumber()])*2*np.pi
det.initialize_pixels_with_beam_information(polar=[0,1,0], wavevector = Wavevector)



# Calculate the pattern
time1= time.time()
pattern_stack = det.read_one_pattern(particle, beam, device = "gpu")

time2 = time.time()
print("Total time usage for the pattern is %f" %(time2- time1))

pattern_2d = det.convert_image_stack_to_2d_image(image_stack=pattern_stack)


np.save('../output/image_stack_'+name+'.npy',pattern_stack)
np.save('../output/image_no_shot_noise_'+name+'.npy', pattern_2d)
```

# Add Noise and Various Detector Effects

```python

# Add detector corrections and shot noise 
diffraction_pattern = np.load('../output/image_stack_'+name+'.npy')

## Modify the distribution
diffraction_pattern = np.multiply(diffraction_pattern, det.polarization_correction) 
diffraction_pattern = np.multiply(diffraction_pattern, det.solid_angle_correction)

## Normalize with respect to the photon number
## The photon number 83005038 is calculate randomly from another image
normalize_factor = (339699. - 135378 )/np.sum(diffraction_pattern)

## Add shot noise
diffraction_pattern_noisy = np.random.poisson(diffraction_pattern* normalize_factor)

# Calculate background noise and convert it to photon number
noise = np.random.uniform( 0, 2* np.sqrt(3*det.pixel_rms))
noise_2d = det.convert_image_stack_to_2d_image(noise)

## cross talk noise
ADU = add_cross_talk_effect_panel('../input/cross_talk_lib.npy', diffraction_pattern_noisy)
ADU_2d = det.convert_image_stack_to_2d_image(ADU)
ADU_2d_noisy = ADU_2d + noise_2d
```

# Downsampling by 4

```python
from skimage.measure import block_reduce
```

```python
patten = block_reduce(ADU_2d_noisy - det.convert_image_stack_to_2d_image(np.sqrt(3*det.pixel_rms)) ,(4,4), np.sum)

fig = plt.figure(figsize=(10,10))

plt.imshow(patten/130.,vmax=10, vmin=0)
plt.show()
```
